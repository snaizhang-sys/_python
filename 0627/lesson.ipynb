```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5, 0.1)
y1 = np.sin(t * np.pi)
y2 = np.cos(t * np.pi) 

figure = plt.figure()
ax1 = figure.add_subplot(2, 1, 1)
ax1.plot(t, y1, linestyle="--", color="red")
ax2 = figure.add_subplot(2, 1, 2)
ax2.plot(t, y2, linestyle="-.",color="blue")
plt.show()
```
