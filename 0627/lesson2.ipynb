```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,
         labels=labels,
         colors=colors,
         explode=[0.3, 0, 0, 0],
         shadow=True,
         autopct="%2.1f%%",
         startangle=180)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 解決中文顯示問題 (在 Windows 環境下通常需要此設定)
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 

labels = ['奶茶', '紅茶', '綠茶', '咖啡']
values = [50, 25, 35, 40]
colors = ['yellow', 'blue', 'red', 'green'] # 修正拼字

figure = plt.figure() 
x = figure.add_subplot()

x.pie(values,
      labels=labels,
      colors=colors,
      explode=[0.3, 0, 0, 0],
      shadow=True,
      autopct="%2.1f%%",
      startangle=180)

plt.show()
```
